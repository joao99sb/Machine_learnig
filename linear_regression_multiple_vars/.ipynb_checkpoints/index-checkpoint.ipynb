# Machine Learning With Python: Linear Regression Multiple Variables

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
from sklearn import linear_model
```

## How to predict the value of a Home based on certain params?

```python
df = pd.read_csv("homeprices.csv")
df = df.drop(df.index[-1])
df
```

# The best value to predict a value with a single variable is to use the mean

```python
import math 

bd_median = math.floor(df.bedrooms.median())
bd_median
```

```python
df.bedrooms = df.bedrooms.fillna(bd_median)
df
```

 We will use regression with multiple variables here. Price can be calculated using following equation,
 

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,3,40]])
```
